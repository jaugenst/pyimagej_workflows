```python
import pandas
from pandas import DataFrame
import tkinter as tk
from tkinter import filedialog
```

```python
root = tk.Tk()
root.withdraw()
dataset = filedialog.askopenfilename()
```

```python
df = pandas.read_csv(dataset)
df.head()
```

```python
ax = df.plot(x='time', y='LV')
```

```python
#ax = df.plot(x='time', y='LV')
df = df.assign(first_derivative=df.diff().eval('LV/time'))
df = df.assign(second_derivative=df.diff().eval('first_derivative/time'))
df.plot(x='time', y='second_derivative')
```

```python
df.plot(x='time', y='first_derivative')
```
